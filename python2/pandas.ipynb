```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("./fietsdiefstal.csv")
df
```

```python
kleuren = Counter(df["Plaats"].fillna('Nan').str.lower())
kleuren
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = []
y = []

for item in reversed(kleuren.most_common(5)):
    x.append(item[0])
    y.append(item[1])

ax.barh(x, y)
plt.show()
```

```python
tmp = df.groupby(['Plaats'])['Plaats'].count().sort_values(ascending=False)
tmp
```

```python
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()
xx, yy = [],[]

for item in tmp.items():
    if item[1] < 20:
        continue
    xx.append(item[0])
    yy.append(item[1])


ax1.barh(xx, yy)
```
