```python
#Starthuur is 20000 euro per jaar met een huurverhoging van 5% per jaar.
#Starthuur is 23000 euro per jaar met een huurverhoging van 3% per ja

def calc_rent(start, perc, years):
    nums = []
    current_price = start
    for year in range(years):
        nums.append(current_price)
        current_price += current_price*perc

    return nums

a = calc_rent(20000, 0.05, 10)
b = calc_rent(23000, 0.03, 10)
print(a)
print(b)

plt.plot(a, label="gebouw a")
plt.plot(b, label="gebouw b")
plt.title("huur")
plt.xlabel('jaar')
plt.ylabel('kosten')
plt.legend()
plt.show()
```
